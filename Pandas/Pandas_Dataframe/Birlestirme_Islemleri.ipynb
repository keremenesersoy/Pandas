```python
import numpy as np
import pandas as pd
```

```python
np_Arr = np.random.randint(1,30,(5,3))
df1 = pd.DataFrame(np_Arr , columns=["var1","var2","var3"])
df2 = pd.DataFrame(np_Arr + 5 , columns=["var1","var2","var3"])
df2
```

### Birleştirilen Dataframede index Düzenleme

```python
pd.concat([df1,df2] , ignore_index=True) # 2 dataframe i alt alta birleştirir # eğer ignore_index True yaparsak indexler düzenlenir
```

```python
df1.columns
```

```python
df2.columns = ["d1","var2","d3"]
df2
```

### Farklı sütunlara sahip Dartaframe leri birleştirme

```python
pd.concat([df1,df2] , ignore_index=True) # eğer birleştirme işleminde farklı sütunlar varsa onlarıda oluşturur ama Nan olarak ayarlanır
```

### Birleştirme işleminde Join Parametresi

```python
pd.concat([df1,df2] , join="inner") # birleştirme esnasında join parametresi inner yazarsak inner join yapar yani sadece eşleşen kolonları birleştirir
```

# İleri Birleştirme İşlemleri

```python
dict1 = {'calisanlar' : ['Ali' , 'Veli' , 'Ayse' , 'Fatma'],
         'grup' : ['Muhasebe' , 'Muhendislik' , 'Muhendislik' , 'IK']
         }
dict2 = {'calisanlar' : ['Ayse' ,'Ali' , 'Veli' , 'Fatma'],
         'ilk_giris' : [2010 , 2009 , 2014 , 2019]}
df1 = pd.DataFrame(dict1)
df2 = pd.DataFrame(dict2)
```

### Merge

```python
pd.merge(df1,df2 , on="calisanlar") # merge fonksiyonu belirli bir ortak sütuna göre birleştirme yapar , on parametresi ile ortak sütun belirleriz
```
